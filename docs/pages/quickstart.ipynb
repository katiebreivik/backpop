# Quickstart

To get started with BackPop, we will first apply it to a simple test case. We will model a binary system composed of two main sequence stars with the following "observables":

$\rm{m_1} = 35 \pm 1 \ M_\odot$

$\rm{m_2} = 5 \pm 0.05 \ M_\odot$

$\rm{porb} = 2500 \pm 50 \ \rm{days}$

$\rm{ecc} = 0.3 \pm 0.02$

## Imports

```python
from backpop import BackPop
from backpop import BackPopsteriors
```

## Ini file

For this run, we will use [simple_SSE.ini](https://github.com/backpop/backpop/tree/main/example_inis). There are four main sections that have been set up in this file:

1. BackPop settings
2. Observational constraints
3. Sampling priors
4. COSMIC-METISSE dictionaries

We go into the details of this file in the [tutorials](tutorials.rst).

## Running BackPop

```python
bp = BackPop(config_file='../../example_inis/simple_SSE.ini')

bp.run_sampler()
```

## View results

```python
posteriors = BackPopsteriors(file='../../example_inis/simple_SSE_output/posteriors.h5')

BPP = posteriors.bpp
kick_info = posteriors.kick_info
```

```python
BPP
```

```python
kick_info
```

```python
fig = posteriors.cornerplot(quantiles=[0.05,0.68,0.95],
                            show_titles=True,
                            smooth=True,
                            plot_contours=True,
                            fill_contours=False,
                            plot_datapoints=False,
                            )
```
