# Quickstart

To get started with ``BackPop``, we will first apply it to a simple test case. <br> <br>
We will model binary system composed of main sequence stars with the following "observables": <br> <br>
$\rm{m_1} = 35 \pm 1 \ M_\odot$ <br>
$\rm{m_2} = 5 \pm 0.05 \ M_\odot$ <br>
$\rm{porb} = 2500 \pm 50 \ \rm{days}$ <br>
$\rm{ecc} = 0.3 \pm 0.02$ <br>
